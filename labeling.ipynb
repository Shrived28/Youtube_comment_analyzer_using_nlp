```python
import re
import pandas as pd
from transformers import pipeline
from tqdm import tqdm
```

```python
df1 = pd.read_csv("-2RAq5o5pwc_user_comments.csv")
df2 = pd.read_csv("3y5A4paFOb4_user_comments.csv")
df3 = pd.read_csv("73_1biulkYk_user_comments.csv")
df4 = pd.read_csv("hkb8oCMrxDg_user_comments.csv")
df5 = pd.read_csv("lC9emrW0F2o_user_comments.csv")
df6 = pd.read_csv("OCvsEfuTA_8_user_comments.csv")
df7 = pd.read_csv("oeCytho8wq4_user_comments.csv")
df8 = pd.read_csv("YbJOTdZBX1g_user_comments.csv")
df9 = pd.read_csv("YR12Z8f1Dh8_user_comments.csv")
df  = pd.DataFrame()
```

```python
df['comments']  = pd.concat([df1['comment'], df2['comment'],df3['comment'], df4['comment'],df5['comment'], df6['comment'],df7['comment'], df8['comment']], axis=0,names=['comments'])
```

```python
df.describe()
```

```python
df
```

```python
classifier = pipeline("sentiment-analysis", model="siebert/sentiment-roberta-large-english")
```

```python

labels = []
for comment in df['comments']:
    try:
        labels.append(classifier(comment)[0]['label'])
    except:
        labels.append(None)
```

```python
labels
```

```python
df['labels'] = labels
```

```python
df.isna().sum()
```

```python
df['labels'].value_counts()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.to_csv("cleaned_labeled_data.csv",index=False)
```
